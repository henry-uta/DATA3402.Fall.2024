```python
def compute_roc_curves_dnn(df, RawNames,FeatureNames, model_function):

    Train_Sample=df[:N_Train]
    Test_Sample=df[N_Train:]
    
    X_Train_r=Train_Sample[RawNames[1:]]
    y_Train_r=Train_Sample["signal"]

    X_Test_r=Test_Sample[RawNames[1:]]
    y_Test_r=Test_Sample["signal"]

    X_Train_f=Train_Sample[FeatureNames[1:]]
    y_Train_f=Train_Sample["signal"]

    X_Test_f=Test_Sample[FeatureNames[1:]]
    y_Test_f=Test_Sample["signal"]

    X_Train_rf=Train_Sample[VarNames[1:]]
    y_Train_rf=Train_Sample["signal"]

    X_Test_rf=Test_Sample[VarNames[1:]]
    y_Test_rf=Test_Sample["signal"]
    
    model_r = model_function(input_dim=X_Train_r.shape[1])
    model_f = model_function(input_dim=X_Train_f.shape[1])
    model_rf = model_function(input_dim=X_Train_rf.shape[1])

    model_r.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model_f.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model_rf.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    model_r.fit(X_Train_r,y_Train_r)
    model_f.fit(X_Train_f,y_Train_f)
    model_rf.fit(X_Train_rf,y_Train_rf)

    y_r = model_r.predict(X_Train_r)
    y_f = model_f.predict(X_Train_f)
    y_rf = model_rf.predict(X_Train_rf)

    # Calculate ROC curve and AUC for training data
    fpr_r, tpr_r, _ = roc_curve(y_Train_r, y_r)
    auc_r = roc_auc_score(y_Train_r, y_r)

    fpr_f, tpr_f, _ = roc_curve(y_Train_f, y_f)
    auc_f = roc_auc_score(y_Train_f, y_f)

    fpr_rf, tpr_rf, _ = roc_curve(y_Train_rf, y_rf)
    auc_rf = roc_auc_score(y_Train_rf, y_rf)

    plt.plot(fpr_r, tpr_r, label='Raw ROC curve (AUC = %0.2f)' % auc_r)
    plt.plot(fpr_f, tpr_f, label='Feature ROC curve (AUC = %0.2f)' % auc_f)
    plt.plot(fpr_rf, tpr_rf, label='Combined ROC curve (AUC = %0.2f)' % auc_rf)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()

    # Determine the best model
    auc_scores = {'model_r': auc_r, 'model_f': auc_f, 'model_rf': auc_rf}
    best_model_name = max(auc_scores, key=auc_scores.get)

    if best_model_name == 'model_r':
        best_model = model_r
    elif best_model_name == 'model_f':
        best_model = model_f
    else:
        best_model = model_rf

    print(f"Best model: {best_model_name} with AUC = {auc_scores[best_model_name]:.2f}")
    return best_model
    
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
best_model_1 = compute_roc_curves_dnn(df, RawNames, FeatureNames, model_1)
```

```python
def compute_max_significance(model, X_test, y_test, n_signal, n_background):
    from sklearn.metrics import roc_curve
    
    # Predict probabilities
    y_pred_proba = model.predict(X_test)
    
    # Compute ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    
    # Compute signal and background counts
    signal_counts = tpr * n_signal
    background_counts = fpr * n_background
    
    # Compute significance
    significance = signal_counts / np.sqrt(signal_counts + background_counts + 1e-9)
    
    # Find the maximum significance and corresponding threshold
    max_significance = np.max(significance)
    optimal_threshold = thresholds[np.argmax(significance)]
    
    print(f"Max Significance: {max_significance:.3f} at Threshold: {optimal_threshold:.3f}")
    return max_significance, optimal_threshold
```
