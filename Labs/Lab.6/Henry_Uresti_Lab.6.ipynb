```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Deck Class
class Deck:
    def __init__(self, sets):
        self.__sets = sets
        self.__hand = ['Spade', 'Club', 'Heart', 'Diamond']
        self.__number = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
        self.plastic_removed = False  # Flag to track if 'Plastic' card has been removed
        self.build_deck()
        self.shuffle(1000)

    def build_deck(self):
        self.totaldeck = []
        for _ in range(self.__sets):
            for number in self.__number:
                for suit in self.__hand:
                    self.totaldeck.append((number, suit))
        if not self.plastic_removed:
            self.totaldeck.append(('Plastic',))  # Include 'Plastic' card if not removed

    def shuffle(self, n):
        for _ in range(n):
            x = random.randint(0, len(self.totaldeck) - 1)
            y = random.randint(0, len(self.totaldeck) - 1)
            self.totaldeck[x], self.totaldeck[y] = self.totaldeck[y], self.totaldeck[x]
        return self.totaldeck

    def draw(self):
        if not self.totaldeck:
            self.build_deck()
            self.shuffle(1000)
        draw = self.totaldeck.pop(0)
        if draw[0] == 'Plastic':
            # Remove 'Plastic' card permanently
            self.plastic_removed = True
            # Rebuild and reshuffle the deck without 'Plastic' card
            self.build_deck()
            self.shuffle(1000)
            # Draw a new card
            return self.draw()
        else:
            return draw

# Strategy Classes
class Strategy:
    def should_hit(self, player_hand, dealer_upcard):
        raise NotImplementedError("This method should be overridden by subclasses.")

class Strategy6(Strategy):
    def should_hit(self, player_hand, dealer_upcard):
        count = 0
        for card in player_hand:
            rank = card[0]
            if rank in ['2', '3', '4', '5', '6']:
                count += 1
            elif rank in ['10', 'Jack', 'Queen', 'King', 'Ace']:
                count -= 1
        return count <= 0

class DealerStrategy(Strategy):
    def should_hit(self, player_hand, dealer_upcard):
        player_hand_value = self.calculate_hand_value(player_hand)
        return player_hand_value < 17

    def calculate_hand_value(self, hand):
        hand_value = 0
        aces = 0
        for card in hand:
            if card[0] == 'Plastic':
                continue  # Skip the 'Plastic' card
            rank = card[0]
            if rank == 'Ace':
                aces += 1
                hand_value += 11
            elif rank in ['Jack', 'Queen', 'King']:
                hand_value += 10
            else:
                hand_value += int(rank)
        while hand_value > 21 and aces:
            hand_value -= 10
            aces -= 1
        return hand_value

# Player Class
class Player:
    def __init__(self, chips, strategy=None, human=True):
        self.__hand = []
        self.chips = chips
        self.bet = 0
        self.__human = human
        self.strategy = strategy

    def draw_player(self, deck):
        self.__hand = []
        for _ in range(2):
            self.__hand.append(deck.draw())

    def is_human(self):
        return self.__human

    def play_human(self, deck):
        pass  # Skipping human input for simulation

    def play_hand(self, dealer_upcard, deck):
        self.bet = 100  # Fixed bet for simulation
        while True:
            player_hand_value = self.calculate_hand_value()
            should_hit = self.strategy.should_hit(self.get_hand(), dealer_upcard)
            if should_hit:
                self.hit(deck)
                player_hand_value = self.calculate_hand_value()
                if player_hand_value > 21:
                    break
            else:
                break

    def get_hand(self):
        return self.__hand

    def hit(self, deck):
        self.__hand.append(deck.draw())

    def calculate_hand_value(self):
        hand_value = 0
        aces = 0
        for card in self.__hand:
            if card[0] == 'Plastic':
                continue  # Skip the 'Plastic' card
            rank = card[0]
            if rank == 'Ace':
                aces += 1
                hand_value += 11
            elif rank in ['Jack', 'Queen', 'King']:
                hand_value += 10
            else:
                hand_value += int(rank)
        while hand_value > 21 and aces:
            hand_value -= 10
            aces -= 1
        return hand_value

# Dealer Class
class Dealer:
    def __init__(self):
        self.__hand = []

    def draw_dealer(self, deck):
        self.__hand = []
        for _ in range(2):
            self.__hand.append(deck.draw())

    def show_hand(self):
        return self.__hand[0]

    def get_hand(self):
        return self.__hand

    def hit(self, deck):
        self.__hand.append(deck.draw())

    def calculate_hand_value(self):
        hand_value = 0
        aces = 0
        for card in self.__hand:
            if card[0] == 'Plastic':
                continue  # Skip the 'Plastic' card
            rank = card[0]
            if rank == 'Ace':
                aces += 1
                hand_value += 11
            elif rank in ['Jack', 'Queen', 'King']:
                hand_value += 10
            else:
                hand_value += int(rank)
        while hand_value > 21 and aces:
            hand_value -= 10
            aces -= 1
        return hand_value

    def play_dealer(self, deck):
        while self.calculate_hand_value() < 17:
            self.hit(deck)

# BJGame Class
class BJGame:
    def __init__(self, decks, verbose=False):
        self.__deck = Deck(decks)
        self.__players = []
        self.verbose = verbose

    def addPlayer(self, player):
        self.__players.append(player)

    def newRound(self):
        dealer = Dealer()
        # Players place their bets and draw initial hands
        for player in self.__players:
            player.draw_player(self.__deck)
        dealer.draw_dealer(self.__deck)

        # Players play their hands
        dealer_upcard = dealer.show_hand()
        for player in self.__players:
            if player.is_human():
                player.play_human(self.__deck)
            else:
                player.play_hand(dealer_upcard, self.__deck)

        # Dealer plays
        dealer.play_dealer(self.__deck)

        # Determine outcomes
        for player in self.__players:
            player_hand_value = player.calculate_hand_value()
            dealer_hand_value = dealer.calculate_hand_value()
            if player_hand_value > 21:
                player.chips -= player.bet
            elif dealer_hand_value > 21 or player_hand_value > dealer_hand_value:
                player.chips += player.bet
            elif player_hand_value < dealer_hand_value:
                player.chips -= player.bet
            else:
                pass  # Push; no chips won or lost

    def play_game(self, rounds=50, strategy_player_index=0):
        round_number = 0
        while round_number < rounds and self.__players[strategy_player_index].chips > 0:
            self.newRound()
            round_number += 1
        # Return the strategy player's final chip count
        return self.__players[strategy_player_index].chips

# Simulate games
def simulate_games(num_games=100, num_rounds=50):
    starting_chips = 1000
    final_chips_list = []

    for _ in range(num_games):
        game = BJGame(decks=6, verbose=False)
        
        strategy6_player = Player(chips=starting_chips, strategy=Strategy6(), human=False)
        game.addPlayer(strategy6_player)
        
        for _ in range(3):
            dealer_strategy_player = Player(chips=starting_chips, strategy=DealerStrategy(), human=False)
            game.addPlayer(dealer_strategy_player)
        
        final_chips = game.play_game(rounds=num_rounds, strategy_player_index=0)
        winnings = final_chips - starting_chips
        final_chips_list.append(winnings)

    return final_chips_list


```
