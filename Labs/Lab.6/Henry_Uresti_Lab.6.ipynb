1. Begin by creating a classes to represent cards and decks. The deck should support more than one 52-card set. The deck should allow you to shuffle and draw cards. Include a "plastic" card, placed randomly in the deck. Later, when the plastic card is dealt, shuffle the cards before the next deal.

```python
import random

class Deck:
    def __init__(self, sets):
        self.__sets = sets
        self.__hand = ['Spade', 'Club', 'Heart', 'Diamond']
        self.__number = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
        self.plastic_removed = False
        self.build_deck()
        self.shuffle(50)

    def build_deck(self):
        self.totaldeck = []
        for _ in range(self.__sets):
            for number in self.__number:
                for suit in self.__hand:
                    self.totaldeck.append((number, suit))
        if not self.plastic_removed:
            self.totaldeck.append(('Plastic', ''))

    def shuffle(self, n):
        if len(self.totaldeck) == 0:
            self.build_deck()
        
        for _ in range(n):
            if len(self.totaldeck) > 1:
                x = random.randint(0, len(self.totaldeck) - 1)
                y = random.randint(0, len(self.totaldeck) - 1)
                self.totaldeck[x], self.totaldeck[y] = self.totaldeck[y], self.totaldeck[x]

    def draw(self):
        while True:
            if not self.totaldeck:
                self.build_deck()
                self.shuffle(50)
            draw = self.totaldeck.pop(0) 
            if draw[0] == 'Plastic':

                self.plastic_removed = True

                self.build_deck()
                self.shuffle(50)

                continue
            else:

                return draw

deck = Deck(2)

print(deck.totaldeck) 
print("===============")
print(deck.shuffle(1000))
print("==============")
print(deck.draw())

```

```python
class HiLoCardCountingStrategy(Strategy):
    def __init__(self):
        self.running_count = 0
        self.cards_seen = 0

    def update_running_count(self, card):
        rank = card[0]
        if rank in ['2', '3', '4', '5', '6']:
            self.running_count += 1
        elif rank in ['10', 'Jack', 'Queen', 'King', 'Ace']:
            self.running_count -= 1
        self.cards_seen += 1

    def should_hit(self, player_hand, dealer_upcard):
        hand_value = self.calculate_hand_value(player_hand)
        dealer_value = self.get_card_value(dealer_upcard[0])

        if hand_value <= 11:
            return True 
        elif hand_value >= 17:
            return False  
        elif 12 <= hand_value <= 16:
            if dealer_value >= 7:
                return True 
            else:
                return False 
        else:
            return False

    def calculate_hand_value(self, hand):
        hand_value = 0
        aces = 0
        for card in hand:
            rank = card[0]
            if rank == 'Ace':
                aces += 1
                hand_value += 11
            elif rank in ['Jack', 'Queen', 'King']:
                hand_value += 10
            else:
                hand_value += int(rank)
        while hand_value > 21 and aces:
            hand_value -= 10
            aces -= 1
        return hand_value

    def get_card_value(self, card_rank):
        if card_rank in ['Jack', 'Queen', 'King']:
            return 10
        elif card_rank == 'Ace':
            return 11
        else:
            return int(card_rank)

    def decide_bet(self, base_bet):

        if self.running_count >= 3:
            return base_bet * 2  
        else:
            return base_bet

class BJGame:
    def __init__(self, decks):
        self.__deck = Deck(decks)
        self.__players = list()

    def addPlayer(self, player):
        self.__players.append(player)

    def newRound(self, verbose=True):
        dealer = Dealer()
        self.__deck.shuffle(1000)
        for player in self.__players:
            if isinstance(player.strategy, HiLoCardCountingStrategy):
                player.bet = player.strategy.decide_bet(10)  # Base bet of $10
            player.draw_player(self.__deck)
        dealer.draw_dealer(self.__deck)

        if verbose:
            print("The upcard is: ", dealer.show_hand())

        if isinstance(player.strategy, HiLoCardCountingStrategy):
            player.strategy.update_running_count(dealer.show_hand())

        for player in self.__players:
            if player.is_human():
                print("Your hand is: ", player.get_hand())
                player.play_human(self.__deck)
            else:
                player.play_hand(dealer.show_hand(), self.__deck)

        dealer.play_dealer(self.__deck)

        for player in self.__players:
            for hand in player.get_hands():
                player_hand_value = player.calculate_hand_value(hand)
                dealer_hand_value = dealer.calculate_hand_value()
                if player_hand_value > 21:
                    if verbose:
                        print("Player busted. Dealer wins.")
                    player.chips -= player.bet
                elif dealer_hand_value > 21 or player_hand_value > dealer_hand_value:
                    if verbose:
                        print("Player wins.")
                    player.chips += player.bet
                elif player_hand_value < dealer_hand_value:
                    if verbose:
                        print("Dealer wins.")
                    player.chips -= player.bet
                else:
                    if verbose:
                        print("Push.")

class Player:
    def __init__(self, chips, strategy=None, human=True):
        self.__hands = [list()]  # List of hands (to handle splits)
        self.chips = chips
        self.bet = 0
        self.__human = human
        self.strategy = strategy

    def is_human(self):
        return self.__human
    
    def draw_player(self, deck):
        self.__hands = [[]]  # Reset hands
        for _ in range(2):
            card = deck.draw()
            self.__hands[0].append(card)
            if isinstance(self.strategy, HiLoCardCountingStrategy):
                self.strategy.update_running_count(card)

    def play_human(self, deck):
        bet = input("Place Bet: ")
        self.bet = int(bet)
        hand = self.__hands[0]  # Play only the first hand initially

        while True:
            print("Your hand is: ", hand)
            if len(hand) == 2 and hand[0][0] == hand[1][0]:
                play = input("Hit, Stand, Double, or Split? ")
            elif len(hand) == 2:
                play = input("Hit, Stand, or Double? ")
            else:
                play = input("Hit or Stand? ")

            if play.lower() == 'hit':
                card = deck.draw()
                hand.append(card)
                if isinstance(self.strategy, HiLoCardCountingStrategy):
                    self.strategy.update_running_count(card)
                hand_value = self.calculate_hand_value(hand)
                if hand_value > 21:
                    print("You busted!")
                    break
            elif play.lower() == 'stand':
                break
            elif play.lower() == 'double' and len(hand) == 2:
                self.bet *= 2
                card = deck.draw()
                hand.append(card)
                if isinstance(self.strategy, HiLoCardCountingStrategy):
                    self.strategy.update_running_count(card)
                print("Your hand is: ", hand)
                break
            elif play.lower() == 'split' and len(hand) == 2 and hand[0][0] == hand[1][0]:
                self.split_hand(deck)
                break
            else:
                print("Invalid input. Please enter a valid action.")

    def play_hand(self, dealer_upcard, deck):
        if self.strategy is None:
            raise ValueError("Strategy is not defined for a non-human player.")

        for hand in self.__hands:
            while True:
                player_hand_value = self.calculate_hand_value(hand)
                should_hit = self.strategy.should_hit(hand, dealer_upcard)
                if should_hit:
                    card = deck.draw()
                    hand.append(card)
                    if isinstance(self.strategy, HiLoCardCountingStrategy):
                        self.strategy.update_running_count(card)
                    if player_hand_value > 21:
                        break
                else:
                    break

    def split_hand(self, deck):

        hand = self.__hands.pop(0)
        card1, card2 = hand
        self.__hands.append([card1, deck.draw()])  
        self.__hands.append([card2, deck.draw()]) 

        if isinstance(self.strategy, HiLoCardCountingStrategy):
            self.strategy.update_running_count(self.__hands[0][1])
            self.strategy.update_running_count(self.__hands[1][1])

        print("Hands after split: ")
        for i, h in enumerate(self.__hands):
            print(f"Hand {i + 1}: {h}")

    def get_hand(self):

        return self.__hands[0]

    def get_hands(self):
 
        return self.__hands

    def hit(self, deck, hand):
        card = deck.draw()
        hand.append(card)
        if isinstance(self.strategy, HiLoCardCountingStrategy):
            self.strategy.update_running_count(card)

    def calculate_hand_value(self, hand):
        hand_value = 0
        aces = 0
        for card in hand:
            if card[0] == 'Ace':
                aces += 1
                hand_value += 11
            elif card[0] in ['Jack', 'Queen', 'King']:
                hand_value += 10
            else:
                hand_value += int(card[0])
        while hand_value > 21 and aces:
            hand_value -= 10
            aces -= 1
        return hand_value

class Dealer:
    def __init__(self):
        self.__hand = list()

    def draw_dealer(self, deck):
        self.__hand = []
        for _ in range(2):
            self.__hand.append(deck.draw())

    def show_hand(self):
        return self.__hand[0]

    def get_hand(self):
        return self.__hand

    def hit(self, deck):
        self.__hand.append(deck.draw())

    def calculate_hand_value(self):
        hand_value = 0
        aces = 0
        for card in self.__hand:
            if card[0] == 'Ace':
                aces += 1
                hand_value += 11
            elif card[0] in ['Jack', 'Queen', 'King']:
                hand_value += 10
            else:
                hand_value += int(card[0])
        while hand_value > 21 and aces:
            hand_value -= 10
            aces -= 1
        return hand_value

    def play_dealer(self, deck):
        while self.calculate_hand_value() < 17:
            self.hit(deck)
```
