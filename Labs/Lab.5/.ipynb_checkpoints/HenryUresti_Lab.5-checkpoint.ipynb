3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n,m,c):
    return [[c for j in range(m)] for i in range(n)]
def zeros(n,m):
    return [[float(0) for j in range(m)] for i in range(n)]
def ones(n,m):
    return [[float(1) for j in range(m)] for i in range(n)]
def eye(n):
    matrix = zeros(n,n)
    for i in range(n):
        matrix[i][i] = 1
    return matrix

print(eye(6))
```

5. Overload python operators to appropriately use your functions in 4 and allow expressions like:
    * 2*M
    * M*2
    * M+N
    * M-N
    * M*N
    * M==N
    * M=N


```python
class matrix:
    def __init__(self, n, m):
        self._n = n
        self._m = m
        self._matrix = [[0 for _ in range(m)] for _ in range(n)]

    def __getitem__(self, index):
        return self._matrix[index]

    def __setitem__(self, index, value):
        self._matrix[index] = value

    def __str__(self):
        return '\n'.join([' '.join(map(str, row)) for row in self._matrix])

    def shape(self):
        return (self._n, self._m)

    def transpose(self):
        transposed = [[self._matrix[i][j] for i in range(self._n)] for j in range(self._m)]
        result = matrix(self._m, self._n)
        result._matrix = transposed
        return result

    def __add__(self, other):
        if not isinstance(other, matrix):
            raise TypeError("Unsupported type for addition.")
        if self.shape() != other.shape():
            raise ValueError("Cannot add matrices of different sizes.")
        result = matrix(self._n, self._m)
        for i in range(self._n):
            for j in range(self._m):
                result[i][j] = self._matrix[i][j] + other[i][j]
        return result

    def __sub__(self, other):
        if not isinstance(other, matrix):
            raise TypeError("Unsupported type for subtraction.")
        if self.shape() != other.shape():
            raise ValueError("Cannot subtract matrices of different sizes.")
        result = matrix(self._n, self._m)
        for i in range(self._n):
            for j in range(self._m):
                result[i][j] = self._matrix[i][j] - other[i][j]
        return result

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            result = matrix(self._n, self._m)
            for i in range(self._n):
                for j in range(self._m):
                    result[i][j] = self._matrix[i][j] * other
            return result
        elif isinstance(other, matrix):
            if self._m != other.shape()[0]:
                raise ValueError("Invalid shapes for matrix multiplication.")
            result = matrix(self._n, other.shape()[1])
            for i in range(self._n):
                for j in range(other.shape()[1]):
                    result[i][j] = sum(self._matrix[i][k] * other[k][j] for k in range(self._m))
            return result
        else:
            raise TypeError("Unsupported type for multiplication. Please use a scalar or another Matrix.")

    def __rmul__(self, other):
        if isinstance(other, (int, float)):
            return self * other
        else:
            raise TypeError("Unsupported type for multiplication.")

    def __eq__(self, other):
        if not isinstance(other, matrix):
            raise TypeError("Unsupported type for comparison.")
        if self.shape() != other.shape():
            return False
        for i in range(self._n):
            for j in range(self._m):
                if self._matrix[i][j] != other[i][j]:
                    return False
        return True
```
