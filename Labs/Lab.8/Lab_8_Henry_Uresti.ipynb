```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import HTML, display
import tabulate

filename = "SUSY.csv"
VarNames=["signal", "l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", 
          "l_2_phi", "MET", "MET_phi", "MET_rel", "axial_MET",
          "M_R", "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]

RawNames=["l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi", "MET", "MET_phi"]
FeatureNames=list(set(VarNames[1:]).difference(RawNames))
df = pd.read_csv(filename, dtype='float64', names=VarNames)
df.head(20)
```

```python
import sklearn.discriminant_analysis as DA
Fisher = DA.LinearDiscriminantAnalysis()
```

```python
N_Train=4000000

Train_Sample=df[:N_Train]
Test_Sample=df[N_Train:]

X_Train=Train_Sample[VarNames[1:]]
y_Train=Train_Sample["signal"]

X_Test=Test_Sample[VarNames[1:]]
y_Test=Test_Sample["signal"]

Test_sig=Test_Sample[Test_Sample.signal==1]
Test_bkg=Test_Sample[Test_Sample.signal==0]
```

```python
Fisher.fit(X_Train,y_Train)
```

```python
plt.figure()
plt.hist(Fisher.decision_function(Test_sig[VarNames[1:]]),bins=100,histtype="step", color="blue", label="signal",stacked=True)
plt.hist(Fisher.decision_function(Test_bkg[VarNames[1:]]),bins=100,histtype="step", color="red", label="background",stacked=True)
plt.legend(loc='upper right')
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

y_train_pred_prob = Fisher.predict_proba(X_Train)[:, 1]

# Predictions on test data
y_test_pred_prob = Fisher.predict_proba(X_Test)[:, 1]

# Calculate ROC curve and AUC for training data
fpr_train, tpr_train, _ = roc_curve(y_Train, y_train_pred_prob)
auc_train = roc_auc_score(y_Train, y_train_pred_prob)

# Calculate ROC curve and AUC for test data
fpr_test, tpr_test, _ = roc_curve(y_Test, y_test_pred_prob)
auc_test = roc_auc_score(y_Test, y_test_pred_prob)

# Plot ROC curves
plt.plot(fpr_train, tpr_train, label='Train ROC curve (AUC = %0.2f)' % auc_train)
plt.plot(fpr_test, tpr_test, label='Test ROC curve (AUC = %0.2f)' % auc_test)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#There seems to be no bias in the training and testing data.
```

```python
def compare_raw_vs_features(df, RawNames, FeatureNames, raw_r, raw_f, raw_rf):

    N_Train=4000000

    Train_Sample=df[:N_Train]
    Test_Sample=df[N_Train:]

    X_Train_r=Train_Sample[RawNames[1:]]
    y_Train_r=Train_Sample["signal"]

    X_Test_r=Test_Sample[RawNames[1:]]
    y_Test_r=Test_Sample["signal"]

    raw_r.fit(X_Train_r,y_Train_r)

    Train_Sample=df[:N_Train]
    Test_Sample=df[N_Train:]

    X_Train_f=Train_Sample[FeatureNames[1:]]
    y_Train_f=Train_Sample["signal"]

    X_Test_f=Test_Sample[FeatureNames[1:]]
    y_Test_f=Test_Sample["signal"]

    raw_f.fit(X_Train_f,y_Train_f)

    Train_Sample=df[:N_Train]
    Test_Sample=df[N_Train:]

    X_Train_rf=Train_Sample[VarNames[1:]]
    y_Train_rf=Train_Sample["signal"]

    X_Test_rf=Test_Sample[VarNames[1:]]
    y_Test_rf=Test_Sample["signal"]

    raw_rf.fit(X_Train_rf,y_Train_rf)

    y_r = raw_r.predict_proba(X_Train_r)[:, 1]
    y_f = raw_f.predict_proba(X_Train_f)[:, 1]
    y_rf = raw_rf.predict_proba(X_Train_rf)[:,1]

    # Calculate ROC curve and AUC for training data
    fpr_r, tpr_r, _ = roc_curve(y_Train_r, y_r)
    auc_r = roc_auc_score(y_Train_r, y_r)

    fpr_f, tpr_f, _ = roc_curve(y_Train_f, y_f)
    auc_f = roc_auc_score(y_Train_f, y_f)

    fpr_rf, tpr_rf, _ = roc_curve(y_Train_rf, y_rf)
    auc_rf = roc_auc_score(y_Train_rf, y_rf)

    plt.plot(fpr_r, tpr_r, label='Raw ROC curve (AUC = %0.2f)' % auc_r)
    plt.plot(fpr_f, tpr_f, label='Feature ROC curve (AUC = %0.2f)' % auc_f)
    plt.plot(fpr_rf, tpr_rf, label='Combined ROC curve (AUC = %0.2f)' % auc_rf)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def compute_maximal_significance(N_S, N_B):
    return N_S / np.sqrt(N_S + N_B)

```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (roc_curve, auc, precision_score, recall_score,
                             f1_score, accuracy_score, confusion_matrix)
from sklearn.model_selection import train_test_split

model = LogisticRegression(max_iter=10000)
model.fit(X_Train, y_Train)

y_pred = model.predict(X_Test)
y_scores = model.predict_proba(X_Test)[:, 1]  # Probability estimates for the positive class

tn, fp, fn, tp = confusion_matrix(y_Test, y_pred).ravel()

tpr = tp / (tp + fn)

fpr = fp / (fp + tn)

fpr_curve, tpr_curve, thresholds = roc_curve(y_Test, y_scores)
roc_auc = auc(fpr_curve, tpr_curve)

precision = precision_score(y_Test, y_pred)
recall = recall_score(y_Test, y_pred)  # Same as TPR
f1 = f1_score(y_Test, y_pred)
accuracy = accuracy_score(y_Test, y_pred)

youden_index = tpr_curve - fpr_curve
max_index = np.argmax(youden_index)
max_significance_threshold = thresholds[max_index]
max_significance = youden_index[max_index]

metrics = {
    'Metric': ['True Positive Rate (TPR)', 'False Positive Rate (FPR)', 'Area Under Curve (AUC)',
               'Precision', 'Recall', 'F1 Score', 'Accuracy', 'Maximal Significance (Youden\'s J)'],
    'Value': [tpr, fpr, roc_auc, precision, recall, f1, accuracy, max_significance]
}

df_metrics = pd.DataFrame(metrics)
print(df_metrics)
```
