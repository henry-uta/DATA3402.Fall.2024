11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, canvas):
        self.canvas = canvas
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def draw(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.draw(self.canvas)
        self.canvas.display()

canvas = Canvas(50, 20)

drawing = RasterDrawing(canvas)

circle = Circle(5, 30, 10)          
rectangle = Rectangle(5, 10, 5, 10) 
triangle = Triangle("Triangle", 5, 5, 15, 15, 25, 5) 

drawing.add_shape(circle)
drawing.add_shape(rectangle)
drawing.add_shape(triangle)

print("Initial Drawing:")
drawing.draw()

drawing.remove_shape(rectangle)
another_circle = Circle(3, 10, 5) 
drawing.add_shape(another_circle)

print("\nModified Drawing:")

drawing.draw()

```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]

    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]

    def line(self, x1, y1, x2, y2, **kwargs):
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        dx = abs(x2 - x1)
        dy = abs(y2 - y1)
        x, y = x1, y1

        sx = -1 if x1 > x2 else 1
        sy = -1 if y1 > y2 else 1

        if dx == 0 and dy == 0:
            self.set_pixel(y, x, **kwargs)
            return

        if dx > dy:
            err = dx / 2.0
            while x != x2:
                self.set_pixel(y, x, **kwargs)
                err -= dy
                if err < 0:
                    y += sy
                    err += dx
                x += sx
        else:
            err = dy / 2.0
            while y != y2:
                self.set_pixel(y, x, **kwargs)
                err -= dx
                if err < 0:
                    x += sx
                    err += dy
                y += sy
        self.set_pixel(y, x, **kwargs)

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def __init__(self, name):
        self.name = name

    def draw(self, canvas, char='*'):
        raise NotImplementedError

    def __repr__(self):
        return f"{self.__class__.__name__}()"

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__("Rectangle")
        self.length = length
        self.width = width
        self.x = x
        self.y = y

    def draw(self, canvas, char='*'):
        for i in range(self.width):
            canvas.set_pixel(self.y, self.x + i, char=char)
            canvas.set_pixel(self.y + self.length - 1, self.x + i, char=char)
        for i in range(self.length):
            canvas.set_pixel(self.y + i, self.x, char=char)
            canvas.set_pixel(self.y + i, self.x + self.width - 1, char=char)

    def __repr__(self):
        return f"Rectangle({self.length}, {self.width}, {self.x}, {self.y})"

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__("Circle")
        self.radius = radius
        self.x = x
        self.y = y

    def draw(self, canvas, char='*'):
        x0, y0 = int(self.x), int(self.y)
        radius = int(self.radius)
        x = radius
        y = 0
        radius_error = 1 - x

        while x >= y:
            points = [
                (x + x0, y + y0),
                (y + x0, x + y0),
                (-x + x0, y + y0),
                (-y + x0, x + y0),
                (-x + x0, -y + y0),
                (-y + x0, -x + y0),
                (x + x0, -y + y0),
                (y + x0, -x + y0),
            ]
            for px, py in points:
                canvas.set_pixel(py, px, char=char)
            y += 1
            if radius_error < 0:
                radius_error += 2 * y + 1
            else:
                x -= 1
                radius_error += 2 * (y - x + 1)

    def __repr__(self):
        return f"Circle({self.radius}, {self.x}, {self.y})"

class Triangle(Shape):
    def __init__(self, name, x1, y1, x2, y2, x3, y3):
        super().__init__(name)
        self.x1, self.y1 = x1, y1
        self.x2, self.y2 = x2, y2
        self.x3, self.y3 = x3, y3

    def draw(self, canvas, char='*'):
        self.safe_line(canvas, self.x1, self.y1, self.x2, self.y2, char)
        self.safe_line(canvas, self.x2, self.y2, self.x3, self.y3, char)
        self.safe_line(canvas, self.x3, self.y3, self.x1, self.y1, char)

    def safe_line(self, canvas, x1, y1, x2, y2, char):
        canvas.line(x1, y1, x2, y2, char=char)

    def __repr__(self):
        return f"Triangle('{self.name}', {self.x1}, {self.y1}, {self.x2}, {self.y2}, {self.x3}, {self.y3})"

class RasterDrawing:
    def __init__(self, canvas, shapes=None):
        self.canvas = canvas
        self.shapes = shapes if shapes else []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def draw(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.draw(self.canvas)
        self.canvas.display()

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(self.__repr__())

    def __repr__(self):
        shapes_repr = ", ".join(repr(shape) for shape in self.shapes)
        canvas_repr = f"Canvas({self.canvas.width}, {self.canvas.height})"
        return f"RasterDrawing({canvas_repr}, [{shapes_repr}])"

def load_raster_drawing(filename):
    with open(filename, "r") as f:
        content = f.read()
    return eval(content)

canvas = Canvas(50, 20)

drawing = RasterDrawing(canvas)

circle = Circle(5, 30, 10)
rectangle = Rectangle(5, 10, 5, 10)
triangle = Triangle("Triangle", 5, 5, 15, 15, 25, 5)

drawing.add_shape(circle)
drawing.add_shape(rectangle)
drawing.add_shape(triangle)

print("Initial Drawing:")
drawing.draw()

filename = "drawing.txt"
drawing.save(filename)
print(f"\nDrawing saved to {filename}")

drawing.shapes = []  
another_circle = Circle(3, 10, 5)
drawing.add_shape(another_circle)

print("\nModified Drawing:")
drawing.draw()

loaded_drawing = load_raster_drawing(filename)
print("\nLoaded Drawing:")
loaded_drawing.draw()
```
